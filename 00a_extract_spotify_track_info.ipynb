```python
import pandas as pd
import os
from spotipy.oauth2 import SpotifyClientCredentials
import json
import spotipy
import time
from os import walk
import signal


from datetime import datetime

from tqdm.notebook import tqdm

SPOTIPY_CLIENT_ID=[]
SPOTIPY_CLIENT_SECRET=[]

n_spotify_clients=len(SPOTIPY_CLIENT_ID)
```

```python
client_index=0
client_credentials_manager = SpotifyClientCredentials(client_id=SPOTIPY_CLIENT_ID[client_index], 
                                                      client_secret=SPOTIPY_CLIENT_SECRET[client_index])
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager, requests_timeout= 5)
```

```python
tracks_df= pd.read_csv(os.path.join('data','ranking_all_201701_202211.csv'), index_col=0)
tracks_df= tracks_df.drop_duplicates('song_id')
```

```python
tracks_df
```

```python
total_track_id_lst= tracks_df['song_id'].unique()
```

```python
filenames = []
for (dirpath, dirnames, f) in walk(os.path.join('data', 'generated', 'spotify')):
    filenames.extend(f)
    break
filenames
```

```python
processed_tracks_df= None
processed_tracks_id_lst= []
if len(filenames)>0:
    df_lst=[]
    for f in filenames:
        df= pd.read_csv(os.path.join('data', 'generated', 'spotify', f), index_col=0)
        df_lst.append(df)
    processed_tracks_df= pd.concat(df_lst, axis=0)    
    processed_tracks_id_lst= processed_tracks_df['track_id'].unique()
```

```python
tracks_id_to_process_lst= list(set(total_track_id_lst) - set(processed_tracks_id_lst))
tracks_id_to_process_lst 
```

```python
len(tracks_id_to_process_lst)
```

```python
track_batch_size= 50
n_batches= len(tracks_id_to_process_lst) // track_batch_size
```

```python
# Register an handler for the timeout
def time_out_handler(signum, frame):
    print("Spotify call timeout!")
    raise Exception("end of time")
    
signal.signal(signal.SIGALRM, time_out_handler)
```

```python
track_info_lst=[]

counter=0
for i in tqdm(range(0, n_batches)):
    counter += 1
    track_ids= tracks_id_to_process_lst[i*track_batch_size:(i+1)*track_batch_size]
    
    signal.alarm(10) # Fire exeption after 10 seconds
    query_successful= True
    try:  
        tracks_info= sp.tracks(track_ids)
    except:
        signal.alarm(0)
        query_successful= False
        if len(track_info_lst)>0:
            print("Saving intermediate results...", end='')       
            now = datetime.now()
            dt_string = now.strftime("%Y%m%d_%H%M%S")
            aux_track_info_df= pd.DataFrame(track_info_lst, columns='track_id artists release_date release_date_precision'.split())
            aux_track_info_df.to_csv(os.path.join('data', 'generated', 'spotify', f'00a_spotify_tracks_{dt_string}.csv'))
            print('DONE')
        else:
            print("No intermediate results to save")
            
        track_info_lst=[]       
        counter=0
        
        #Refresh client to use for the extraction
        client_index = (client_index + 1) % n_spotify_clients
        print(f"Using new client  #{client_index}/{n_spotify_clients} ({SPOTIPY_CLIENT_ID[client_index]}, {SPOTIPY_CLIENT_SECRET[client_index]})")
        client_credentials_manager = SpotifyClientCredentials(client_id=SPOTIPY_CLIENT_ID[client_index], 
                                                              client_secret=SPOTIPY_CLIENT_SECRET[client_index])
        
        sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager, requests_timeout= 5)
    
    #Cancel alarm if spotify call return successfully
    signal.alarm(0)
    if query_successful:
        for track_info in tracks_info['tracks']:
            try:
                track_artists_lst=[]
                for artist in track_info['artists']:
                    track_artists_lst.append((artist['id'], artist['name']))
                release_date= track_info['album']['release_date']
                release_date_precision= track_info['album']['release_date_precision']
                track_info_lst.append((track_info['id'], track_artists_lst, release_date, release_date_precision))
            except:
                print("Error retrieving song data")
```

```python
now = datetime.now()
dt_string = now.strftime("%Y%m%d_%H%M%S")
aux_track_info_df= pd.DataFrame(track_info_lst, columns='track_id artists release_date release_date_precision'.split())
aux_track_info_df.to_csv(os.path.join('data', 'generated', 'spotify', f'00a_spotify_tracks_{dt_string}.csv'))
```
