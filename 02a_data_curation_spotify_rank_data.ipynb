### Read artist genres

```python
artist_genres_df= pd.read_csv(os.path.join('data','artists_genres_2017_2022.csv'), index_col=0)
artist_genres_df
```

```python
artist_genres_df['name']= artist_genres_df['name'].fillna(artist_genres_df['0'])
artist_genres_df = artist_genres_df[artist_genres_df['name'].notna()].copy()

artist_genres_df['name']= artist_genres_df['name'].apply(lambda x: x.strip())
artist_genres_df= artist_genres_df.drop(columns='0')
```

```python
artist_genres_df
```

```python
genres_counts= None
target_cols= artist_genres_df.columns.drop('name')
for c in target_cols:
    if genres_counts is None:
        genres_counts= artist_genres_df[c].value_counts()
    else:
        genres_counts= genres_counts.add(artist_genres_df[c].value_counts(), fill_value=0)

genres_counts= genres_counts.sort_values(ascending=False)        
genres_counts
```

```python
genres_counts[:10].index.tolist()
```

```python
target_genres= genres_counts.index.tolist()
```

```python
final_artist_genres=[]
for artist_id, artist_info in artist_genres_df.iterrows():
    artist_name= artist_info['name'].strip()
    artist_genres= []
    for c in target_cols:
        if artist_info[c] in target_genres:
            artist_genres.append(artist_info[c])
    final_artist_genres.append((artist_id, artist_name, len(artist_genres), artist_genres))        
```

```python
final_artist_genres_df= pd.DataFrame(final_artist_genres, columns='id name n_genres genres'.split())
final_artist_genres_df= final_artist_genres_df[final_artist_genres_df['n_genres']>0]
final_artist_genres_df
```

```python
final_artist_genres_df.to_csv(os.path.join('data', 'generated', '02_artist_genres.csv'))
```

### Map artist names with codes

```python
artists_map={}
for i,r in final_artist_genres_df.iterrows():
    artists_map[r['name'].strip()]= r['id'].strip()
```

```python
target_cols= ['Artist', 'Artist_2', 'Artist_3', 'Artist_4', 'Artist_5', 'Artist_6',
              'Artist_7', 'Artist_8', 'Artist_9', 'Artist_10', 'Artist_11',
              'Artist_12', 'Artist_13', 'Artist_14', 'Artist_15', 'Artist_16',
              'Artist_17', 'Artist_18', 'Artist_19', 'Artist_20', 'Artist_21',
              'Artist_22', 'Artist_23', 'Artist_24', 'Artist_25', 'Artist_26',
              'Artist_27', 'Artist_28', 'Artist_29', 'Artist_30', 'Artist_31',
              'Artist_32', 'Artist_33', 'Artist_34', 'Artist_35', 'Artist_36',
              'Artist_37', 'Artist_38', 'Artist_39', 'Artist_40']

collab_w_codes_df= collab_df.copy()
for col in target_cols:
    collab_w_codes_df[col]=collab_w_codes_df[col].apply(lambda x: artists_map.get(x.strip(), 'no_code'))
```

```python
def count_artists_w_code(row):
    counter=0
    for col in target_cols:
        if row[col] != 'no_code':
            counter+= 1
    return counter

collab_w_codes_df['n_artists']= collab_w_codes_df.apply(lambda x: count_artists_w_code(x), axis=1)    
```

```python
collab_w_codes_df= collab_w_codes_df[collab_w_codes_df['n_artists']>0].reset_index(drop=True)
```
