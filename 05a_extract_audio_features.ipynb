```python
client_index=0
client_credentials_manager = SpotifyClientCredentials(client_id=SPOTIPY_CLIENT_ID[client_index], 
                                                      client_secret=SPOTIPY_CLIENT_SECRET[client_index])
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager, requests_timeout= 5)
```

### Retrieve all target tracks

```python
all_tracks= pd.read_csv(os.path.join('data', 'generated', '04_lfm_spotify_tracks_w_features.csv'), index_col=0)
tracks_to_process= all_tracks[all_tracks['danceability'].isna()]
tracks_to_process= tracks_to_process.set_index('track_id')
tracks_to_process
```

### Read tracks that have been already processed

```python
filenames = []
for (dirpath, dirnames, f) in walk(os.path.join('data', 'generated', 'audio_features')):
    filenames.extend(f)
    break
    
processed_tracks_df= None
processed_tracks_id_lst= []
if len(filenames)>0:
    df_lst=[]
    for f in filenames:
        df= pd.read_csv(os.path.join('data', 'generated', 'audio_features', f), index_col=0)
        df_lst.append(df)
    processed_tracks_df= pd.concat(df_lst, axis=0)    
    processed_tracks_id_lst= processed_tracks_df['id'].unique()
len(processed_tracks_id_lst)
```

```python
tracks_id_to_process_lst= list(set(tracks_to_process.index.tolist())-set(processed_tracks_id_lst))
len(tracks_id_to_process_lst)
```

```python
track_batch_size= 100
n_batches= len(tracks_id_to_process_lst) // track_batch_size
```

```python
# Register an handler for the timeout
def time_out_handler(signum, frame):
    print("Spotify call timeout!")
    raise Exception("end of time")
    
signal.signal(signal.SIGALRM, time_out_handler)
```

```python
track_feat_lst=[]

for i in tqdm(range(0, n_batches)):
    track_ids= tracks_id_to_process_lst[i*track_batch_size:(i+1)*track_batch_size]
    
    signal.alarm(10) # Fire exeption after 10 seconds
    query_successful= True
    try:  
        features_lst= sp.audio_features(track_ids)
    except:
        signal.alarm(0)
        query_successful= False
        if len(track_feat_lst)>0:
            print("Saving intermediate results...", end='')       
            now = datetime.now()
            dt_string = now.strftime("%Y%m%d_%H%M%S")
            aux_track_info_df= pd.DataFrame(track_feat_lst)
            aux_track_info_df.to_csv(os.path.join('data', 'generated', 'audio_features', f'05_audio_features_{dt_string}.csv'))
            print('DONE')
        else:
            print("No intermediate results to save")
            
        track_feat_lst=[]       
        
        #Refresh client to use for the extraction
        client_index = (client_index + 1) % n_spotify_clients
        print(f"Using new client  #{client_index}/{n_spotify_clients} ({SPOTIPY_CLIENT_ID[client_index]}, {SPOTIPY_CLIENT_SECRET[client_index]})")
        client_credentials_manager = SpotifyClientCredentials(client_id=SPOTIPY_CLIENT_ID[client_index], 
                                                              client_secret=SPOTIPY_CLIENT_SECRET[client_index])
        
        sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager, requests_timeout= 5)
    
    #Cancel alarm if spotify call return successfully
    signal.alarm(0)
    if query_successful:
        for track_features in features_lst:
            if track_features:
                track_feat_lst.append(track_features)
```

```python
aux_track_info_df= pd.DataFrame(track_feat_lst)
aux_track_info_df.head()
```

```python
now = datetime.now()
dt_string = now.strftime("%Y%m%d_%H%M%S")
aux_track_info_df.to_csv(os.path.join('data', 'generated', 'audio_features', f'05_audio_features_{dt_string}.csv'))
```
